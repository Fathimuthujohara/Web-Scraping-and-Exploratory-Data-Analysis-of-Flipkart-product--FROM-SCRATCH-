```python
soup=BeautifulSoup(html_text,'lxml')
```

```python
product=soup.find_all("div",class_="_4ddWXP")
```

```python
print(product)
```

```python
len(product)
```

```python
box=product[1]
box
```

```python
title=box.find("a",class_="s1Q9rs").text.strip()
price=box.find("div",class_="_30jeq3").text.strip()
ratings=box.find("div",class_="_3LWZlK").text.strip()
weight_of_the_product=box.find("div",class_="_3Djpdu").text.strip()
offers=box.find("div",class_="_3Ay6Sb").text.strip()
sponsored=box.find("div",class_="_4HTuuX").text.strip()
print(title,price,ratings,weight_of_the_product,offers,sponsored)
#print(title)
```

```python
all_products=[]
for box in product:
    temp_dict={}
    temp_dict['Name of the sunscreen']=box.find('a',class_="s1Q9rs").text.strip()
    temp_dict['Price']=box.find("div",class_="_30jeq3").text.replace("₹","").strip()
    temp_dict['Ratings']=box.find("div",class_="_3LWZlK")
    temp_dict['Ratings']=temp_dict['Ratings'].get_text(strip=True) if temp_dict['Ratings'] else '-'
    temp_dict['Weight']=box.find("div",class_="_3Djpdu").text.strip()
    temp_dict['Offers']=box.find("div",class_="_3Ay6Sb")
    temp_dict['Offers']=temp_dict['Offers'].get_text(strip=True) if temp_dict['Offers'] else '-'
    temp_dict['Sponsored']=box.find("div",class_="_4HTuuX")
    temp_dict['Sponsored']=temp_dict['Sponsored'].get_text(strip=True) if temp_dict['Sponsored'] else '-'
    all_products.append(temp_dict)
```

```python
df=pd.DataFrame(all_products)
df
```

```python
df.to_csv("Sunscreen.csv",index=False,encoding='utf-8')
```
